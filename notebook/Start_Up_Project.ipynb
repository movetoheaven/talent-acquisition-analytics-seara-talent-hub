## Business Question 3
### How ready are candidates to support startup workforce needs?
1. Pengkategorian skill level
2. Kesiapan kandidat

### Variabel:

1. Jumlah Skill
2. Competency Level

### Output:

1. Emerging Talent
2. Job-Ready Talent
3. Multi-Skilled Talent

### Visual:

Competency Level Distribution

```python
def competency_level(row):
  excel = row['Excel']
  sql = row['SQL']
  python = row['Python']
  power_bi = row['Power BI']
  looker = row['Looker Studio']
  tableau = row['Tableau']
  db = row['Databases']

  total = excel + sql + python + power_bi + looker + tableau + db

  # High-Potential Talent
  if total >=5:
    return 'High-Potential Talent'

  # Job-Ready Talent
  elif total >= 3:
    return 'Job-Ready Talent'

  # Emerging
  else:
    return 'Emerging Talent'

df['Competency Level'] = df.apply(competency_level, axis=1)

komptensi = df['Competency Level'].value_counts()

plt.figure(figsize=(6,5))

komptensi.plot(kind='bar')

plt.ylabel('Jumlah Kandidat')
plt.title('Kategori Skill Kandidat')
plt.xticks(rotation=0)

plt.show()
```

```python
pd.crosstab(
    df['Competency Level'],
    df['Mahasiswa/Fresh Graduate']
).plot(kind='bar', stacked=True)

plt.title('Kategori Skill vs Mahasiswa/Fresh Graduate')
plt.show()
```

```python
# @title
pd.crosstab(
    df['Competency Level'],
    df['Mahasiswa/Fresh Graduate']
)

# 0 means not mhs/fresh grad, 1 means mhs/freshgrad
```

```python
source_competency = pd.crosstab(
    df['Mengetahui Perusahaan ini Dari Mana?'],
    df['Competency Level']
)

source_competency

source_competency.plot(
    kind='bar',
    stacked=True,
    figsize=(10,6)
)

plt.title('Candidate Competency Level by Recruitment Source')
plt.xlabel('Recruitment Source')
plt.ylabel('Number of Candidates')
plt.legend(title='Competency Level')
plt.show()
```

```python
motivation_count = df[motivation_cols].sum().sort_values(ascending=False)

plt.figure(figsize=(7,5))
motivation_count.plot(kind='bar')

plt.ylabel('Jumlah Kandidat')
plt.title('Motivasi Bergabung')
plt.xticks(rotation=20)
plt.show()
```

```python
motivations = [
    'Cari Kerja',
    'Upskill',
    'Networking',
    'Switch Career'
]

motivation_fg = pd.DataFrame()

for motive in motivations:
    motivation_fg[motive] = df.groupby(
        'Mahasiswa/Fresh Graduate'
    )[motive].sum()

motivation_fg
```

```python
motivation_fg.T.plot(
    kind='bar',
    figsize=(10,6)
)

plt.title('Candidate Motivation by Fresh Graduate Status')
plt.xlabel('Motivation')
plt.ylabel('Number of Candidates')
plt.legend(title='Fresh Graduate')

plt.show()
```
